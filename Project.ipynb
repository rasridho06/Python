```python
import random

# Ini adalah gambar-gambar yang akan kita tampilkan. Anggap saja ini adalah "stiker" pilihan kita.
rock = ''' (gambar batu) '''
paper = ''' (gambar kertas) '''
scissors = ''' (gambar gunting) '''

# Kita kumpulkan stiker-stiker ini ke dalam sebuah "album stiker".
# Posisi stikernya: album_stiker[0] itu batu, album_stiker[1] itu kertas, album_stiker[2] itu gunting.
album_stiker = [rock, paper, scissors]

# --- Meminta Pilihan Pemain (dan Pastikan Benar) ---

# Anggap ini seperti mengisi formulir. Kita minta angka pilihan.
# Kita pakai "satpam" (try-except) untuk jaga-jaga kalau yang diisi bukan angka.
try:
    pilihan_kamu = int(input("Pilih angka: 0=Batu, 1=Kertas, 2=Gunting "))
except ValueError:
    print("Aduh, inputnya salah! Masukin angka ya, bukan yang lain.")
    exit() # Langsung bubar kalau salah input.

# Kalau angkanya sudah benar (misal: 0, 1, 2), baru lanjut ke bawah.
# Tapi, kita masih butuh "penjaga gerbang" (if) untuk cek apakah angkanya 0, 1, atau 2.
if pilihan_kamu < 0 or pilihan_kamu > 2:
    print("Wah, angkanya di luar jangkauan (harus 0, 1, atau 2). Kamu kalah deh!")
else:
    # --- Pilihan Komputer ---
    # Komputer ngundi acak dari 0 sampai 2.
    pilihan_komputer = random.randint(0, 2)

    # --- Tampilkan Pilihan ---
    print("\nKamu memilih:")
    print(album_stiker[pilihan_kamu]) # Ambil stiker dari album sesuai pilihanmu

    print("\nKomputer memilih:")
    print(album_stiker[pilihan_komputer]) # Ambil stiker dari album sesuai pilihan komputer

    # --- Menentukan Pemenang (Aturan Main) ---
    if pilihan_kamu == pilihan_komputer:
        print("Seri! Main lagi yuk.")
    # Ini aturan kalau kamu menang:
    # (Batu vs Gunting) ATAU (Kertas vs Batu) ATAU (Gunting vs Kertas)
    elif (pilihan_kamu == 0 and pilihan_komputer == 2) or \
         (pilihan_kamu == 1 and pilihan_komputer == 0) or \
         (pilihan_kamu == 2 and pilihan_komputer == 1):
        print("Yeay! Kamu menang!")
    # Kalau bukan seri dan kamu juga tidak menang, berarti kamu kalah.
    else:
        print("Yah, kamu kalah deh.")
```

```python
import random

rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

# Ini adalah list yang berisi semua gambar.
# Kita bisa mengakses gambar dengan indeks: game_images[0] = rock, game_images[1] = paper, game_images[2] = scissors
game_images = [rock, paper, scissors]

# Meminta input dari pengguna dan mengkonversinya ke integer
try:
    user_choice = int(input("Pilih angka: 0 untuk Batu, 1 untuk Kertas, dan 2 untuk Gunting "))
except ValueError:
    print("Input tidak valid. Harap masukkan angka (0, 1, atau 2).")
    exit() # Menghentikan program jika input bukan angka

# Validasi input pengguna
if user_choice < 0 or user_choice > 2:
    print("Angka yang Anda masukkan salah. Anda kalah!")
else:
    # Pilihan komputer secara acak
    pc_choice = random.randint(0, 2)

    # --- BAGIAN BARU: MENAMPILKAN GAMBAR PILIHAN ---
    print("\n------------------------------") # Garis pemisah untuk kejelasan
    print("Pilihan Anda:")
    print(game_images[user_choice]) # Kita gunakan user_choice sebagai indeks untuk mengambil gambar dari list game_images

    print("\nPilihan Komputer:")
    print(game_images[pc_choice]) # Kita gunakan pc_choice sebagai indeks untuk mengambil gambar dari list game_images
    print("------------------------------\n") # Garis pemisah

    # Logika penentuan pemenang (tetap sama)
    if user_choice == pc_choice:
        print("Seri!")
    elif (user_choice == 0 and pc_choice == 2) or \
         (user_choice == 1 and pc_choice == 0) or \
         (user_choice == 2 and pc_choice == 1):
        print("Anda Menang!")
    else:
        print("Anda Kalah!")
```

```python
import random

# Gambar batu, kertas, gunting
rock = '''
_______
---' ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
_______
---' ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
_______
---' ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

# Simpan semua pilihan dalam list agar bisa diakses lewat index
choices = [rock, paper, scissors]

# Ambil input dari user
user_input = input("Please input your number (0 for Rock, 1 for Paper, 2 for Scissors): ")

# Cek apakah input valid (harus angka 0, 1, 2)
if not user_input.isdigit() or int(user_input) not in [0, 1, 2]:
    print("Invalid input. You lose by default.")
else:
    user_choice = int(user_input)
    pc_choice = random.randint(0, 2)

    print(f"\nYou chose:\n{choices[user_choice]}")
    print(f"Computer chose:\n{choices[pc_choice]}")

    # Aturan menang, kalah, seri
    if user_choice == pc_choice:
        print("Draw!")
    elif (user_choice == 0 and pc_choice == 2) or \
         (user_choice == 1 and pc_choice == 0) or \
         (user_choice == 2 and pc_choice == 1):
        print("You win!")
    else:
        print("You lose.")
```
