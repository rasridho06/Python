```python
x = 1 
while x < 4 : #batas pengulangan adalah kurang dari 4, Code a while loop that keeps running as long as x < 4
    print(x)
    x = x+ 0.5 #x=1, 1+0,5 hingga nilainya kurang dari 4, maks di 3,5 karena jika ditambahkan 0,5 lagi maka nilainya akan menjadi 4
```

```python
# Initialize offset
offset=8

# Code the while loop
while offset != 0 : #perulangan akan berhenti jika offset sudah mencapai 0
    print("correcting...")
    offset = offset - 1 
    print(offset)
```

```python
# Initialize offset
offset = -6

# Code the while loop
while offset != 0 : #perulangan akan berhenti jika offset sudah mencapai 0
    print("correcting...")
    if offset > 0 : #jika nilai offset >0 maka akan dikurangi 1
      offset = offset - 1
    else : 
      offset = offset + 1 #jika nilai offset <0 maka akan ditambahkan 1
    print(offset)
```

```python
# areas list
areas = [11.25, 18.0, 20.0, 10.75, 9.50]

# Change for loop to use enumerate() and update print()
for index, area in enumerate(areas) :
    print("room "+str(index)+":"+str(area))
```

```python
# areas list
areas = [11.25, 18.0, 20.0, 10.75, 9.50]

# Code the for loop
for index, area in enumerate(areas) :
    print("room " + str(index+1) + ": " + str(area)) #menambahkan angka satu agar no urut room dimulai dari angka 1, 
    #karena index selalu dimulai dari 0
```

```python
# house list of lists
house = [["hallway", 11.25], 
         ["kitchen", 18.0], 
         ["living room", 20.0], 
         ["bedroom", 10.75], 
         ["bathroom", 9.50]]
         
# Build a for loop from scratch
for room in house:
    print("the " + room[0] + " is " + str(room[1]) + " sqm")
    #room[0] = list index ke 0 yaitu dari hallway hingga bathroom, kolom urut ke 0 yaitu hallway ke bawah
    #room[1] = list index ke 1 yaitu area dari 11.25 sampai 9.5, kolom urut ke 1 yaitu 11.25 - 9.50
```

```python
# Definition of dictionary
europe = {'spain':'madrid', 'france':'paris', 'germany':'berlin',
          'norway':'oslo', 'italy':'rome', 'poland':'warsaw', 'austria':'vienna' }
 
# Iterate over europe
for key, value in europe.items() : #key urutan pertama, #value urutan kedua, .items digunakan untuk memisahkan urutan pertama dan kedua
    print('the capital of ' + key + ' is ' + str(value))
```
